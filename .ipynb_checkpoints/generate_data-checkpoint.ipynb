This program generates spin configurations of the Ising model and stores them into files. Additionaly some statistical calculations are performed to validate the corectness of the Ising Model Class.

```python
%matplotlib inline

import numpy as np
import math
import os
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import timeit
import time

import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
# constants 
T_c = 2 / (math.log(1 + math.sqrt(2)))

cpath = os.getcwd()
ppath = os.path.abspath(os.path.join(cpath, os.pardir))
```

#### Ising Model Class

```python
class IsingModel:
    
    def __init__(self, n = 5, J = 1, kbT = 1, h=0):
        self.n = n
        self.J = J
        self.kbT = kbT
        self.h = h
        self.state = np.ones((n*n), dtype = 'int')
        self.state[0] = -1 #trick for animation: makes the python image object know that there can be two different values.
        self.magnetization = np.sum(self.state)
        
    def get_current_state(self, one_dim = False):
        if one_dim == True:
            return self.state
        else:
            return np.reshape(self.state,(self.n, self.n), 'C')
    
    def get_next_state(self, no_passes, one_dim = False):
        self._update_configuration(no_passes)
        if one_dim == True:
            return self.state
        else:
            return np.reshape(self.state,(self.n, self.n), 'C')
    
    def get_magnetization(self, mag_type = 'rel'):
        self.magnetization = np.sum(self.state)
        if mag_type == 'rel':
            return self.magnetization/(self.n*self.n)
        elif mag_type == 'abs':
            return self.magnetization
        else:
            print('error: wrong magnetization type')
            

#### getter:        
    def get_kbT(self):
        return self.kbT       

    def get_J(self):
        return self.J      

    def get_h(self):
        return self.h     

    def get_n(self):
        return self.n      

    
#### setter:
    def set_kbT(self,kbT):
        self.kbT = kbT        
        
    def set_J(self,J):
        self.J = J
        
    def set_h(self,h):
        self.h = h
        
        
    def set_state(self,state):
        self.state= np.reshape(state,(self.n*self.n))
        
    def reset_model(self):
        self = self.__init__(n = self.n, J = self.J, kbT = self.kbT, h=self.h)

        
#######################################################################
# private functions

    def _update_site(self):
        #updates one single site of the lattice
        #choose a random site (spin) within the lattice
        random_spin = np.random.randint((self.n*self.n))
        rand_row = int(random_spin/self.n)
        rand_col = random_spin % self.n

        row_above = rand_row-1
        row_below = rand_row+1
        col_left = rand_col-1
        col_right = rand_col+1

        #periodic boundary conditions:
        if rand_row == 0:
            row_above = self.n-1
        elif rand_row == (self.n-1):
            row_below = 0

        if rand_col == 0:
            col_left = self.n-1
        elif rand_col == (self.n-1):
            col_right = 0

        # calculate delta_e as the energy after flip minus the energy before flip
        delta_e = self.J * 2 * self.state[random_spin] * \
                (self.state[row_above*self.n + rand_col] + self.state[row_below*self.n + rand_col] + \
                 self.state[rand_row*self.n + col_left] + self.state[rand_row*self.n + col_right]) + \
                (2*self.state[random_spin] * self.h) #term for external magnetic field
        
        
        
        #always flip for delta_e<=0, 
        #flip for delta_e>0 if r<exp(-delta_e / (J*kbT)) --- stimmt nicht??
        if delta_e <= 0:
            self.state[random_spin] *= (-1) 
        else:
            if self._random_bolzmann(delta_e):
                self.state[random_spin] *= (-1) 
                
    def _random_bolzmann(self, delta_e):
        r = np.random.random()
        if r < math.exp(- delta_e / (self.kbT)) :
            return True
        else:
            return False    

    
    def _update_configuration(self, no_passes):
        # updates np_passes sites
        for i in range(self.n*self.n*no_passes):
            self._update_site()
        
        
########################################################################
# grafical functions
    def plot_state(self):
        fig = plt.figure(figsize=(7,7))
        ax = plt.axes()
        image = ax.imshow(self.get_current_state())
        plt.title('Ising Model, n=' + str(self.n) + ' kbT=' + str(self.kbT) + ' J=' +str(self.J))
        plt.show()
            
    def animate(self,i):
        no_passes = 1

        self.image.set_data(self.get_current_state())
        self.get_next_state(no_passes)
        return self.image
    
    def create_animation(self,frames):
        self.frames = frames
        fig = plt.figure(figsize=(7,7))
        ax = plt.axes()
        
        self.image = ax.imshow(self.get_current_state())
        plt.title('Ising Model, n=' + str(self.n) + ' kbT=' + str(self.kbT) + ' J=' +str(self.J))

        #create animation
        self.anim = FuncAnimation(fig, func=self.animate, frames=frames, interval=50)
        return self.anim

    def save_animation(self,title):
        #save as .mp4 file 
        self.anim.save(title, fps=self.frames/20)
```

## Generating Data Files

This code creates new spin configurations from the Ising Model Class and stores them into files. 
The files will be stored in a subsubfolder of the parrent path with the following name: <br>
parrent path -> spin_configurations -> set1 <br>
If someone wishes to generate a second round of data, simply change the variable "set_no" to 2 and run the program again. The resulting files will then be stored in: <br>
parrent path -> spin_configurations -> set2 <br>

The number of generated sets has to be in consecutive order! E.G. 1,2,3,... Otherwise it is not possible to run the perform_spilt.ipynb script.

```python
# variables to set:
n=40
J=1
h=0
kbT_sequence = [1.5, 1.7, 1.8, 1.9, 2.0, 2.1, 2.2, T_c, 2.3, 2.4, 2.5, 2.6, 2.7, 2.8, 3.0, 3.5]

set_no = 1 #number of the folder in which the files will be stored. 

no_configurations = 1000 #number of spin configurations that will be generated per temperature
initial_passes = 500 #number of Monte Carlo passes before the first spin configuration is stored (to reach equilibrium)
distance_passes = 20 #number of MC passes between each spin configuration
```

```python

n_total = n*n
myIsingModel= IsingModel(n=n, J=J, kbT=kbT_sequence[0], h=h)

foldername = "spin_configurations/set"+str(set_no) 

for j,kbT in enumerate(kbT_sequence):
    
    filename = "spin_configurations_"+str(n)+"_"+str(kbT)+".txt"
    filepath = os.path.join(ppath, foldername,filename)
    header= '# This file contains '+str(no_configurations)+' configurations of the Ising model at n='+str(n)+', kbT='+str(kbT)+', J='+str(J)+' and h='+str(h)+'. \n' 
    header+= '# generated with '+str(initial_passes)+' initial passes and '+str(distance_passes)+' passes between each configuration \n'
   
    with open(filepath, 'w') as file:
        file.write(header)
    
    configurations = np.zeros((no_configurations, n_total), dtype='int')
    myIsingModel.reset_model()
    myIsingModel.set_kbT(kbT)
    myIsingModel.get_next_state(initial_passes-distance_passes)
    
    for i in range(no_configurations):
    
        configurations[i] = myIsingModel.get_next_state(distance_passes, True)
        
    configurations = np.where(configurations == -1, 0, configurations)

    print('save file no. '+str(j+1)+'/'+str(len(kbT_sequence)))
    with open(filepath, "a") as file:
        np.savetxt(file, configurations, fmt='%i', delimiter=',')

print('finished')
```

## Data Statistics
In this code sections some calculations are performed to validate the correctness of the Ising Model Class. For this purpose new spin configurations are generated and used in the calculations (not the spin configurations stored in the files).

#### Simulation Class

```python
class Simulation:
    
    def __init__(self, no_iterations, no_passes, ising_model, initial_passes = 500):
        
        self.ising_model = ising_model
        self.no_iterations = no_iterations # over how many state configurations the mag/susc should be calculated
        self.no_passes = no_passes # how many passes should lie between two valid state configurations
        self.initial_passes = initial_passes  # calculate first x passes, to make sure that equilibrium has been reached
        
    def get_no_iterations(self):
        return self.no_iterations
        
    def set_ising_model(self,ising_model):
        self.ising_model = ising_model
        
    def set_no_iterations(self,no_iterations):
        self.no_iterations = no_iterations
        
    def set_no_passes(self,no_passes):
        self.no_passes = no_passes

    def simulate_magnetization_over_T(self, sequence ):
        
        mag_means = np.zeros(len(sequence),dtype='float')
        
        for j,param in enumerate(sequence):
            
            self.ising_model.reset_model()
            self.ising_model.set_kbT(param)
            self.ising_model.get_next_state(self.initial_passes) # to make sure that equilibrium has been reached

            mag_mean = 0
            mag_mean += np.absolute(self.ising_model.get_magnetization('rel'))
            
            for i in range(1,self.no_iterations):

                self.ising_model.get_next_state(self.no_passes) # calculate next 10 passes
                mag_mean += np.absolute(self.ising_model.get_magnetization('rel'))

            mag_mean /= self.no_iterations 
            mag_means[j] = mag_mean
            
            print("kbT: " + str(param) +"\t <|M|>:"+ str(mag_mean))
        
        return mag_means
    
    
    def simulate_magnetization_over_h(self, sequence ):
        
        mag_means = np.zeros(len(sequence),dtype='float')
        
        for j,param in enumerate(sequence):
            
            self.ising_model.reset_model()
            self.ising_model.set_h(param)
            self.ising_model.get_next_state(self.initial_passes) # to make sure that equilibrium has been reached

            mag_mean = 0
            mag_mean += (self.ising_model.get_magnetization('rel'))
            
            for i in range(1,self.no_iterations):

                self.ising_model.get_next_state(self.no_passes) # calculate next 10 passes
                mag_mean += (self.ising_model.get_magnetization('rel'))

            mag_mean /= self.no_iterations 
            mag_means[j] = mag_mean
            
            print("h: " + str(param) +"\t <M>:"+ str(mag_mean))
        
        return mag_means
    
    def simulate_susceptibility_over_kbT(self, sequence):
        
        susc = np.zeros(len(sequence),dtype='float')
        
        for j,param in enumerate(sequence):
            
            self.ising_model.reset_model()
            self.ising_model.set_kbT(param)
            self.ising_model.get_next_state(self.initial_passes)  # to make sure that equilibrium has been reached
            
            current_mag = self.ising_model.get_magnetization('rel') # no absolute value here
            mag = current_mag
            mag_sq = current_mag ** 2
            
            for i in range(1,self.no_iterations):

                self.ising_model.get_next_state(self.no_passes) # calculate next x passes
                current_mag = self.ising_model.get_magnetization('rel')
                mag += current_mag
                mag_sq += current_mag ** 2

            mag = mag / self.no_iterations
            mag_sq = mag_sq / self.no_iterations
            susc[j] = ((mag_sq - mag**2) / param) * pow(self.ising_model.get_n(),2)  # (<m^2> - <m>^2) * (N/kbT) 

            print(str(param) +"\t "+ str(susc[j]))
            
        return susc
    
    
    

    def check_valid_parameter(self, variable_parameter):
        allowed_parameters = ['kbT', 'J', 'h', 'n']
        if not (variable_parameter in allowed_parameters):
            raise ValueError(str(variable_parameter)+" is not a valid parameter.")
            
    def check_valid_measure(self, measure):
        allowed_parameters = ['susc', 'mag']
        if not (measure in allowed_parameters):
            raise ValueError(str(measure)+" is not a valid measure.")
    
    
        
   
```

#### Data Statistics

```python
# this code calculates the mean magnetization over a sequence of kbT values with a fixed external field

kbT_sequence = [1.5, T_c, 3.5]
max_iteration = 5000
mags = np.ones((len(kbT_sequence),max_iteration))
for i in range(len(kbT_sequence)):
    n = 40
    J = 1
    h = 0
    kbT = kbT_sequence[i]
    myIsingModel = IsingModel(n=n, J=J, kbT=kbT, h=h)
    for j in range(max_iteration):

        mags[i,j] = myIsingModel.get_magnetization('rel')
        myIsingModel.get_next_state(1)
```

```python
# this code prints the results from above

x_values = np.arange(max_iteration)

sns.set(font_scale=1.3)
fig = plt.figure(figsize=(7,7))
ax = plt.axes()
ax.set(xlabel='no. passes after initialization', ylabel='magnetization', ylim=(-1,1), xlim=(x_values[0], x_values[-1]))
line1 = ax.plot(x_values, mags[0], label='kbT='+str(kbT_sequence[0])) #, marker='o' , marker='o'
line2 = ax.plot(x_values, mags[1], label='kbT=2.269')
line3 = ax.plot(x_values, mags[2], label='kbT='+str(kbT_sequence[2]))
ax.legend()
ax.grid(True)
plt.title('Magnetization over Time')
plt.show()
```

```python
# this code calculates the mean magnetization for different temperatures and different magnetic fields

kbT_sequence = np.arange(1, 11, 1)

h_sequence = [1,2,10]
mag_means = np.ones((len(h_sequence),len(kbT_sequence)))
no_configurations = 200
no_passes = 10
no_initial_passes = 500
for j in range(len(h_sequence)):

    for i in range(len(kbT_sequence)):
        n = 40
        J = 1
        h = h_sequence[j]
        kbT = kbT_sequence[i]
        myIsingModel = IsingModel(n=n, J=J, kbT=kbT, h=h)
        myIsingModel.get_next_state(no_initial_passes)
        
        for k in range(no_configurations):
            myIsingModel.get_next_state(no_passes)
            mag_means[j,i] += np.absolute(myIsingModel.get_magnetization('rel'))
            
        mag_means[j,i] /= no_configurations
        print("h:"+str(h_sequence[j])+"\t kbT: "+str(kbT_sequence[i])+"\t <|m|>:"+ str(mag_means[j,i]))
        
```
